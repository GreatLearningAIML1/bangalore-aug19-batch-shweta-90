### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
Iris.dtypes

non_floats = []
for col in Iris:
    if Iris[col].dtypes != "float64":
        non_floats.append(col)
Iris_Float = Iris.drop(columns=non_floats)
Iris_Float.describe()


for column in Iris_Float:
    print("Range for Column "+ column  + " is as below:  ")
    print(Iris_Float[column].max() - Iris_Float[column].min())
   
    

    
    
```

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
HR_Data.describe()
First_Quartile = np.quantile(HR_Data['Age'],.25) 
Third_Quartile = np.quantile(HR_Data['Age'],.75)
IQR = Third_Quartile - First_Quartile
print(IQR)
```
